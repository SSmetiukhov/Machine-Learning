```python
import pandas as pd

# details for beer dataset
training_file = "wildfires_training.csv"
test_file = "wildfires_test.csv"
independent_cols = ["year","temp","humidity","rainfall","drought_code","buildup_index","day","month","wind_speed"]
dependent_col = "fire"

# Here we load our training dataset in from the training file using the pandas library
df_training = pd.read_csv(training_file)
print(df_training.head())
print(df_training.shape)
```

```python
# set up a matrix X containing the independent variables from the training data
X_training = df_training.loc[:,independent_cols]
print(X_training.head())
print(X_training.shape)
```

```python
# Set up a vector y containing the dependent variable / target attribute for the training data
y_training = df_training.loc[:,dependent_col]
print(y_training.head())
print(y_training.shape)
```

```python
# Next we load our test dataset in from the file wildfires_test.csv
df_test = pd.read_csv(test_file)
print(df_test.head())
print(df_test.shape)
```

```python
# set up a matrix X containing the independent variables from the test data
X_test = df_test.loc[:,independent_cols]
print(X_test.head())
print(X_test.shape)
```

```python
# Set up a vector y containing the dependent variable  for the test data
y_test = df_test.loc[:,dependent_col]
print(y_test.head())
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn import metrics

# create a model using the default settings
model = DecisionTreeClassifier()
model.fit(X_training, y_training)

predictions_training = model.predict(X_training)
predictions_test = model.predict(X_test)

# compute the accuracy on the training and test set predictions
accuracy_training = metrics.accuracy_score(y_training, predictions_training)
accuracy_test = metrics.accuracy_score(y_test, predictions_test)
print("Accuracy on training data:",accuracy_training)
print("Accuracy on test data:",accuracy_test)

plt.figure(figsize=(8,8))
tree.plot_tree(model)
#tree graph 1
```

```python
criterions = ["gini", "entropy"]
print(criterions)
accuracy_training_i = []
accuracy_test_i = []
for criterion in criterions:
    model_i = DecisionTreeClassifier(criterion=criterion).fit(X_training, y_training)

    # compute the accuracy on the training and test set predictions
    predictions_training_i = model_i.predict(X_training)
    predictions_test_i = model_i.predict(X_test)

    # compute the accuracy on the training and test set predictions
    accuracy_training_i.append(metrics.accuracy_score(y_training, predictions_training_i))
    accuracy_test_i.append(metrics.accuracy_score(y_test, predictions_test_i))
    
    plt.figure(figsize=(8,8))
    tree.plot_tree(model_i)

print(accuracy_training_i)
print(accuracy_test_i)
#tree graph 2
```

```python
import matplotlib.pyplot as plt
plt.scatter(criterions,accuracy_training_i,marker="x")
plt.scatter(criterions,accuracy_test_i,marker="+")
plt.xlim([0, len(criterions)])
plt.ylim([0.0, 1.1])
plt.xlabel("Criterions for the Decidion Tree Classifier")
plt.ylabel("Accuracy")
legend_labels = ["Training","Test"]
plt.legend(labels=legend_labels, loc=4, borderpad=1)
plt.title("Effect of criterion on training and test set accuracy", fontsize=10)
plt.show()
#test 1 graph
```

```python
splitters = ["best", "random"]
accuracy_training_k_n = []
accuracy_test_k_n = []
for j in range (2):
    accuracy_training_k_n.append([])
    accuracy_test_k_n.append([]) 
    for criterion in criterions:
        model_k_n = DecisionTreeClassifier(criterion=criterion, splitter=splitters[j]).fit(X_training, y_training)
        model_k_n.fit(X_training, y_training)

        # compute the accuracy on the training and test set predictions
        predictions_training_k_n = model_k_n.predict(X_training)
        predictions_test_k_n = model_k_n.predict(X_test)

        # compute the accuracy on the training and test set predictions
        accuracy_training_k_n[j].append(metrics.accuracy_score(y_training, predictions_training_k_n))
        accuracy_test_k_n[j].append(metrics.accuracy_score(y_test, predictions_test_k_n))
        
        plt.figure(figsize=(8,8))
        tree.plot_tree(model_k_n)
        
        
    print("splitter =",splitters[j],"training",accuracy_training_k_n[j],"\n")
    print("splitter =",splitters[j],"test",accuracy_test_k_n[j],"\n")
```

```python
accuracy_training_random_gini = []
accuracy_test_random_gini = []
for i in range (1000):
    model_x = DecisionTreeClassifier(criterion="gini", splitter="random").fit(X_training, y_training)

    # compute the accuracy on the training and test set predictions
    predictions_training_x = model_x.predict(X_training)
    predictions_test_x = model_x.predict(X_test)

    # compute the accuracy on the training and test set predictions
    accuracy_training_random_gini.append(metrics.accuracy_score(y_training, predictions_training_x))
    accuracy_test_random_gini.append(metrics.accuracy_score(y_test, predictions_test_x))
    

accuracy_training_random_entropy = []
accuracy_test_random_entropy = []
for i in range (1000):
    model_x = DecisionTreeClassifier(criterion="entropy", splitter="random").fit(X_training, y_training)

    # compute the accuracy on the training and test set predictions
    predictions_training_x = model_x.predict(X_training)
    predictions_test_x = model_x.predict(X_test)

    # compute the accuracy on the training and test set predictions
    accuracy_training_random_entropy.append(metrics.accuracy_score(y_training, predictions_training_x))
    accuracy_test_random_entropy.append(metrics.accuracy_score(y_test, predictions_test_x))
    
    
```
